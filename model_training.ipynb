### Preprocessing

```python
dt_col=[" arrival_time","dep_time","date_of_journey"]
num_col=["duration","total_stops"]
cat_col=[col for col in x_train.columns if (col not in dt_col)and (col not in num_col)]
```

```python
cat_col
```

```python
num_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="mean")),  # Corrected 'medium' to 'mean'
    ("scaler", StandardScaler())
])

cat_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(sparse_output=False,handle_unknown='ignore'))
])

doj_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ('extractor', DatetimeFeatures(features_to_extract=["month", "week", "day_of_week", "day_of_month"], format="mixed")),
    ("scaler", StandardScaler())
])

time_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ('extractor', DatetimeFeatures(features_to_extract=["hour", "minute"], format="mixed")),
    ("scaler", StandardScaler())
])
```

```python
preprocessor=ColumnTransformer(transformers=[
    ('num',num_transformer,num_col),
    ('cat',cat_transformer,cat_col),
    ("doj",doj_transformer,["date_of_journey"]),
    ("time_",time_transformer,["arrival_time","dep_time"])
])
```

```python
preprocessor.fit_transform(x_train)
```

### Model Selection

```python
algorithms={
    'LinearRegressor':LinearRegression(),
    'SupportVectorMachine':SVR(),
    'RandomForest':RandomForestRegressor(n_estimators=10),
    'xgboost':XGBRegressor(n_estimators=10)
}
```

```python
data=pd.concat([train_df,val_df],axis=0)
data
```

```python
x_data,y_data=split(data)
```

```python
print(x_data.shape,y_data.shape)
```

```python
def plot_curve(sizes,mean_score,std_score,label,ax):
    ax.plot(
        sizes,
        mean_score,
        marker='o',
        label=label
    )
    ax.fill_between(
        x=sizes,
        y1=mean_score-std_score,
        y2=mean_score+std_score,
        alpha=.5
    )
```

```python
def plot_learning_curves(name, algorithm, figsize=(12, 4)):
    model = Pipeline(steps=[
        ('pre', preprocessor),
        ('alg', algorithm)
    ])
    
    # Corrected train_sizes parameter
    train_size, train_score, test_score = learning_curve(
        estimator=model,
        X=x_data,
        y=y_data,
        scoring='r2',
        cv=3,
        n_jobs=-1,
        train_sizes=np.linspace(0.1, 1.0, 42)  # Generates 42 points between 0.1 and 1.0
    )
    
    mean_train_score = np.mean(train_score, axis=1)
    std_train_score = np.std(train_score, axis=1)
    train_score_str = f"{mean_train_score[-1]:.2f} +/- {std_train_score[-1]:.2f}"
    
    mean_test_score = np.mean(test_score, axis=1)
    std_test_score = np.std(test_score, axis=1)
    test_score_str = f"{mean_test_score[-1]:.2f} +/- {std_test_score[-1]:.2f}"

    fig, ax = plt.subplots(figsize=figsize)
    
    # Training curve
    plot_curve(
        train_size,
        mean_train_score,
        std_train_score,
        f"train({train_score_str})",
        ax
    )
    
    # Test curve
    plot_curve(
        train_size,
        mean_test_score,
        std_test_score,
        f"test({test_score_str})",
        ax
    )
    
    ax.set(xlabel='Train set size', ylabel='R2 score', title=name)
    plt.legend(loc='lower right')
    plt.show()
```

```python
for name,alg in algorithms.items():
    plot_learning_curves(name,alg)
```

### Model Training

```python
model=Pipeline(steps=[
    ('pre',preprocessor),
    ('rf',RandomForestRegressor(n_estimators=10))
])
```

```python
model.fit(x_data,y_data)
```

### Model Evaluating

```python
def evaluate_model(X, y):
	y_pred = model.predict(X)
	return r2_score(y, y_pred)
```

```python
print(f"R2 score on training data is {evaluate_model(x_data,y_data)}")
```
